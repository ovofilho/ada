```python
def menu_inicial():
    opcao_invalida = True
    lista_opcoes = [0,1,2,3,4]
    while opcao_invalida:
        print('=' * 18, 'Menu Principal', '=' * 18)
        print('''
        1 - Cadastrar restaurante/cardápio
        2 - Listar restaurante/cardápio
        3 - Fazer pedido
        4 - Consultar pedido
        0 - Sair
        ''')
        print('=' *50)
        opcao = int(input('Digite uma opção: '))
        opcao_invalida = opcao not in lista_opcoes
        if opcao_invalida:
            print("Voce digitou uma opcao invalida")
    return opcao
```

```python
def menu_1():
   #opcao_valida = opcao > 0 and opcao <= 4
    opcao_invalida = True
    lista_opcoes = [0,1,2,3,4]
    while opcao_invalida:
        print('=' * 18, 'Menu Cadastrar', '=' * 18)
        print('''
        1 - Cadastrar restaurante
        2 - Cadastar cardápio
        0 - Voltar
        ''')
        print('=' *50)
        opcao1 = int(input('Digite uma opção: '))
        opcao_invalida = opcao1 not in lista_opcoes
        if opcao_invalida:
            print("Voce digitou uma opcao invalida")
    return opcao1
```

```python
def func_cadastrar_restaurante(restaurantes):
    contador = len(restaurantes)
    opcao_invalida = True
    print('=' * 50)
    print('Digite o nome do restaurante a ser adicionado')
    print('=' * 50)
    restaurante_adicionado = input()
    restaurantes.append([contador+1, restaurante_adicionado])
    print(f'O restaurante {restaurante_adicionado} foi adicionado com sucesso\nDeseja adicionar mais algum restaurante?')
    while opcao_invalida:
        print('=' * 50)
        print('''
        1 - Sim
        0 - Não
      ''')
        opcao = int(input('Digite uma opção: '))
        opcao_invalida = opcao not in [1,0]
        if opcao_invalida:
            print("Voce digitou uma opcao invalida")
        if opcao == 1:
            func_cadastrar_restaurante(restaurantes)
```

```python
def func_cadastrar_cardapio(cardapio):
    opcao_invalida = True
    codigo_restaurante = int(input('Codigo do restaurante'))
    
    Teste = False
    for i in restaurantes:
        if codigo_restaurante == i[0]:
            Teste = True

    if Teste:
        nome_do_produto = input('Nome produto: ')
        preco_produto = input('Preco Produto: ')
        tempo_preparo = input('Tempo preparo: ')
        cardapio.append([codigo_restaurante,[nome_do_produto, preco_produto, tempo_preparo]])
    else:
        print("O codigo do de restaurante é invalido")
    while opcao_invalida:
        print('\n')
        print('=' * 50)
        print("Deseja cadastrar outro produto?")
        print('=' * 50)
        print('''
        1 - Sim
        0 - Não
      ''')
        opcao = int(input('Digite uma opção: '))
        opcao_invalida = opcao not in [1,0]
        if opcao_invalida:
            print("Voce digitou uma opcao invalida")
        if opcao == 1:
            func_cadastrar_cardapio(cardapio)
            
```

```python
def func_listar_restaurantes():
    print('=' * 50)
    opcao_invalida = True
    for i in restaurantes:
        print(f"Codigo do restaurante: {i[0]}\nNome do restaurante: {i[1]}\n") 
    print('=' * 50)
    codigo_restaurante_ = int(input("Digite o código do restaurante para listar o cardapio: "))
    print('=' * 50)
    Teste = False
    listagem_cardapio = []
    for i in restaurantes:
        if codigo_restaurante_ == i[0]:
            Teste = True
    listagem_cardapio = []
    for i in cardapio:
        if i[0] == codigo_restaurante_:
            listagem_cardapio.append(i[1])
    if listagem_cardapio == []:
            print("\nCardapio não encontrado")
    
    if Teste:
        for i in listagem_cardapio:
            print(f"\nNome do produto: {i[0]}\nPreço: {i[1]}\nTempo: {i[2]}\n")
            print('=' * 50)
    else:
        print('Codigo invalido')

    while opcao_invalida:
        print('\n')
        print('=' * 50)
        print("Deseja tentar novamente?")
        print('=' * 50)
        print('''
        1 - Sim
        0 - Não
      ''')
        opcao = int(input('Digite uma opção: '))
        opcao_invalida = opcao not in [1,0]
        if opcao_invalida:
            print("Voce digitou uma opcao invalida")
        if opcao == 1:
            func_listar_restaurantes()
```

```python
def func_fazer_pedido():
    opcao_invalida = True
    print('\n')
    print('=' * 50)
    for i in restaurantes:
        print(f"Codigo do restaurante: {i[0]}\nNome do restaurante: {i[1]}\n") 
    print('=' * 50)
    contador = int(len(pedido))+1
    print("Escolha um restaurante")
    codigo_restaurante_ = int(input("Digite o código do restaurante: "))
    print('=' * 50)
    Teste = False
    for i in restaurantes:
        if codigo_restaurante_ == i[0]:
            Teste = True
    if not Teste:
        print("Restaurante não encontrado")
    while Teste:
        listagem_cardapio = []
        for i in cardapio:
            if i[0] == codigo_restaurante_:
                listagem_cardapio.append(i[1])
        if listagem_cardapio == []:
            print("\nEsta esmpre não possui um cardapio cadastrado")
            Teste = False
            break
        for i in listagem_cardapio:
            print(f"\nNome do produto: {i[0]}\nPreço: {i[1]}\nTempo: {i[2]}\n")
            print('=' * 50)
        nome_produto_escolhido = input("Digite o nome do produto: ")
        Teste2 = False
        pedido_escolhido = []
        for i in listagem_cardapio:
            if nome_produto_escolhido == i[0]:
                pedido_escolhido = i
                Teste2 = True
        if Teste2:
            pedido.append([contador, [codigo_restaurante_] + pedido_escolhido])
            print("Item adicionado a lista de pedidos")
            print("\nO numero do seu pedido é:", contador)
            
            Teste = False
        else:
            print("Item não encontrado")
            Teste = False
            
    while opcao_invalida:
        print('\n')
        print('=' * 50)
        print("Deseja tentar novamente?")
        print('=' * 50)
        print('''
        1 - Sim
        0 - Não
      ''')
        opcao = int(input('Digite uma opção: '))
        opcao_invalida = opcao not in [1,0]
        if opcao_invalida:
            print("Voce digitou uma opcao invalida")
        if opcao == 1:
            func_fazer_pedido()
        
    
```

```python
def func_consultar_pedido():
    opcao_invalida = True
    pedido_consultar = int(input("Digite o codigo do pedido: "))
    print('\n')
    print('=' * 50)
    Teste = True
    printar_pedido = False
    
    for i in pedido:
        if pedido_consultar == i[0]:
            printar_pedido = i
            Teste = False
                            
    if Teste:
        print("Pedido não encontrado")
        
    if not Teste:
        print(f"\nCodigo do pedido: {printar_pedido[0]}\nCodigo do restaurante: {printar_pedido[1][0]}\
        \nNome do pedido: {printar_pedido[1][1]}\nPreço do pedido: {printar_pedido[1][2]}\
        \nTempo de preparo: {printar_pedido[1][3]}")
    
    while opcao_invalida:
            print('\n')
            print('=' * 50)
            print("Deseja tentar novamente?")
            print('=' * 50)
            print('''
            1 - Sim
            0 - Não
          ''')
            opcao = int(input('Digite uma opção: '))
            opcao_invalida = opcao not in [1,0]
            if opcao_invalida:
                print("Voce digitou uma opcao invalida")
            if opcao == 1:
                func_consultar_pedido()



```
